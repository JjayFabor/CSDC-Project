```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import math
```

```python
house_df = pd.read_csv("housing.csv")
house_df.head()
```

```python
house_df.corr()
```

```python
dummies = pd.get_dummies(house_df.ocean_proximity)
dummies
```

```python
merged = pd.concat([house_df, dummies], axis='columns')
merged
```

```python
final_house_df = merged.drop(['ocean_proximity', 'NEAR OCEAN'], axis='columns')
final_house_df
```

```python
final_house_df.corr()
```

```python
# Set the missing values of the total_bedrooms to its median
median_total_bedrooms = math.floor(house_df['total_bedrooms'].median())
final_house_df.total_bedrooms = final_house_df['total_bedrooms'].fillna(median_total_bedrooms)
```

```python
features = final_house_df[['housing_median_age', 'total_rooms', 'median_income', '<1H OCEAN', 'NEAR BAY']]
X = features
y = final_house_df.median_house_value
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_train,y_train)
```
